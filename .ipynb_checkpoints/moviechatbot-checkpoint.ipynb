```python
import pandas as pd

df = pd.read_csv("data/archive/movies_metadata.csv", low_memory=False)

df.head()

df.shape
df.columns
df.info()
```

```python
df = df[['title','genres','release_date','vote_average','overview','popularity']]
df['year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year
df.rename(columns={'vote_average': 'rating'}, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['overview','release_date','vote_average'], inplace=True)
```

```python
df['year'] = pd.to_datetime(df['release_date']).dt.year
```

```python
import ast

def clean_genres(x):
    try:
        return " ".join([i['name'] for i in ast.literal_eval(x)])
    except:
        return ""

df['genres_clean'] = df['genres'].apply(clean_genres)
```

```python
df['combined'] = df['title'] + " " + df['overview'] + " " + df['genres_clean']
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

movie_embeddings = model.encode(df['combined'], show_progress_bar=True)
```

```python
#retrieval system - similarity search
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def search_movies(query):
    query_embedding = model.encode([query])
    
    similarity = cosine_similarity(query_embedding, movie_embeddings)
    
    top_indices = np.argsort(similarity[0])[-5:][::-1]
    
    return df.iloc[top_indices]
```
